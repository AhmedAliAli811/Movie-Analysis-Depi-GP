```python
import pandas as pd 
import json
```

```python
movies_df = pd.read_csv('tmdb_5000_movies.csv')
credits_df = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies_normalized = movies_df[['id', 'budget', 'homepage', 'original_language', 'original_title',
                               'overview', 'popularity', 'release_date', 'revenue',
                               'runtime', 'status', 'tagline', 'title', 'vote_average',
                               'vote_count']].rename(columns={'id': 'movie_id'})
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x: json.loads(x) if pd.notnull(x) else [])

movies_with_genres = movies_df[['id', 'genres']].explode('genres').reset_index(drop=True)
movies_with_genres['genre_id'] = movies_with_genres['genres'].apply(lambda x: x['id'] if isinstance(x, dict) else None)
movies_with_genres['genre_name'] = movies_with_genres['genres'].apply(lambda x: x['name'] if isinstance(x, dict) else None)

genres_df = movies_with_genres[['genre_id', 'genre_name']].drop_duplicates().reset_index(drop=True)

movie_genres_df = movies_with_genres[['id', 'genre_id']].rename(columns={'id': 'movie_id'}).drop_duplicates().reset_index(drop=True)
```

```python
movie_genres_df.dropna(inplace=True)
```

```python
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(lambda x: json.loads(x) if pd.notnull(x) else [])
movies_with_languages = movies_df[['id', 'spoken_languages']].explode('spoken_languages').reset_index(drop=True)
movies_with_languages['language_id'] = movies_with_languages['spoken_languages'].apply(lambda x: x['iso_639_1'] if isinstance(x, dict) else None)
movies_with_languages['language_name'] = movies_with_languages['spoken_languages'].apply(lambda x: x['name'] if isinstance(x, dict) else None)
languages_df = movies_with_languages[['language_id', 'language_name']].drop_duplicates().reset_index(drop=True)
movie_languages_df = movies_with_languages[['id', 'language_id']].rename(columns={'id': 'movie_id'}).drop_duplicates().reset_index(drop=True)
```

```python
movies_df['production_companies'] = movies_df['production_companies'].apply(lambda x: json.loads(x) if pd.notnull(x) else [])

movies_with_companies = movies_df[['id', 'production_companies']].explode('production_companies').reset_index(drop=True)
movies_with_companies['company_id'] = movies_with_companies['production_companies'].apply(lambda x: x['id'] if isinstance(x, dict) else None)
movies_with_companies['company_name'] = movies_with_companies['production_companies'].apply(lambda x: x['name'] if isinstance(x, dict) else None)

companies_df = movies_with_companies[['company_id', 'company_name']].drop_duplicates().reset_index(drop=True)
movie_companies_df = movies_with_companies[['id', 'company_id']].rename(columns={'id': 'movie_id'}).drop_duplicates().reset_index(drop=True)

```

```python
movies_df['production_countries'] = movies_df['production_countries'].apply(lambda x: json.loads(x) if pd.notnull(x) else [])
movies_with_countries = movies_df[['id', 'production_countries']].explode('production_countries').reset_index(drop=True)
movies_with_countries['country_iso'] = movies_with_countries['production_countries'].apply(lambda x: x['iso_3166_1'] if isinstance(x, dict) else None)
movies_with_countries['country_name'] = movies_with_countries['production_countries'].apply(lambda x: x['name'] if isinstance(x, dict) else None)
countries_df = movies_with_countries[['country_iso', 'country_name']].drop_duplicates().reset_index(drop=True)
movie_countries_df = movies_with_countries[['id', 'country_iso']].rename(columns={'id': 'movie_id'}).drop_duplicates().reset_index(drop=True)
```

```python
credits_df['crew'] = credits_df['crew'].apply(lambda x: json.loads(x) if pd.notnull(x) else [])

movies_with_crew = credits_df[['movie_id', 'crew']].explode('crew').reset_index(drop=True)
movies_with_crew['crew_id'] = movies_with_crew['crew'].apply(lambda x: x['id'] if isinstance(x, dict) else None)
movies_with_crew['crew_name'] = movies_with_crew['crew'].apply(lambda x: x['name'] if isinstance(x, dict) else None)
movies_with_crew['job'] = movies_with_crew['crew'].apply(lambda x: x['job'] if isinstance(x, dict) else None)
movies_with_crew['department'] = movies_with_crew['crew'].apply(lambda x: x['department'] if isinstance(x, dict) else None)

crew_df = movies_with_crew[['crew_id', 'crew_name']].drop_duplicates().reset_index(drop=True)

jobs_df = movies_with_crew[['job']].drop_duplicates().reset_index(drop=True)
jobs_df['job_id'] = jobs_df.index + 1  

departments_df = movies_with_crew[['department']].drop_duplicates().reset_index(drop=True)
departments_df['department_id'] = departments_df.index + 1

movie_crew_jobs_df = movies_with_crew.merge(jobs_df, on='job').merge(departments_df, on='department')[
    ['movie_id', 'crew_id', 'job_id', 'department_id']]
```

```python
import requests
import pyodbc

```

```python
def fetch_movie_by_id(movie_id, api_key):
    movie_url = f"https://api.themoviedb.org/3/movie/{movie_id}"
    params = {
        'api_key': api_key,
        'language': 'en-US'
    }
    
    response = requests.get(movie_url, params=params)
    
    if response.status_code == 200:
        return response.json()  # Return the movie data in JSON format
    else:
        return None  # Handle error cases



def update_movie_in_db(conn, movie_id, title, release_date, budget, revenue):
    cursor = conn.cursor()
    update_query = """
    UPDATE movies
    SET budget = ?, revenue = ?
    WHERE movie_id = ?;
    """
    cursor.execute(update_query, (budget, revenue, movie_id))
    conn.commit()
    cursor.close()
```
